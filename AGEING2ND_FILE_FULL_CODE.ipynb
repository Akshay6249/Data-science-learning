```python
import pandas as pd

# Specify the path to your input Excel file
input_excel_path = '/content/drive/MyDrive/CURRENT STOCK & AGEING STORE WISE.xlsx'

# Read all sheets from the input Excel file into a dictionary of DataFrames
xlsx = pd.ExcelFile(input_excel_path)
sheet_data = {}

for sheet_name in xlsx.sheet_names:
    df = pd.read_excel(xlsx, sheet_name)
    sheet_data[sheet_name] = df

# Create a new DataFrame by concatenating all the sheets
merged_data = pd.concat(sheet_data.values(), ignore_index=True)

# Specify the path and name for the output Excel file
output_excel_path = '/content/drive/MyDrive/AGEING 09-24-23.xlsx'

# Write the merged data to a new Excel file
merged_data.to_excel(output_excel_path, index=False)

print(f"All sheets merged and saved to '{output_excel_path}'.")
```

```python
# Load Excel data into a pandas DataFrame
file_path = '/content/drive/MyDrive/AGEING 09-24-23.xlsx'  # Update with your file path
df = pd.read_excel(file_path)

# Display the DataFrame
print(df.head())
```

```python
# Create a duplicate DataFrame
df_duplicate = df.copy()

# Display the original DataFrame
print("Original DataFrame:")
print(df.head())

# Display the duplicate DataFrame
print("\nDuplicate DataFrame:")
print(df_duplicate.head())
```

```python
# Specify the number of characters to remove for each column
characters_to_remove = {'DEPT_NAME': 4, 'SEC_NAME': 6, 'CAT_NAME': 8}

# Iterate through the columns and remove the specified number of characters in the copy
for column, num_chars in characters_to_remove.items():
    df_duplicate[column] = df_duplicate[column].str.slice(num_chars)

# Display the updated DataFrame copy
print(df_duplicate.head())
```

```python
unique_values = df_duplicate['DEPT_NAME'].unique()
print(unique_values)
```
